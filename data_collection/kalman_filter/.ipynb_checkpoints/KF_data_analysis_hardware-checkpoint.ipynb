```python

#!/usr/bin/env python3

# read text files and plot them

import matplotlib.pyplot as plt
import numpy as np
import sys
from scipy import signal, fftpack

data_file = "data_file_03-15-19_22-03-27"
path = "/home/elenakern/Codes/sai2.0/applications/inertial_params_application/data_collection/kalman_filter/kalman_filter/"
path_data_file = path + data_file
file = np.loadtxt(path_data_file,skiprows=1)
#print(file)
vel_kin = file[1::,0:3]
acc_sensor = file[1::,3:6]
acc_kin = file[1::,6:9]
kf_vel = file[1::,9:12]
kf_acc = file[1::,12:15]

time = np.arange(np.size(vel_kin[:,0]))
time = time/1000


#data_file_03-15-19_21-09-10
#Q.diagonal() << 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-6, 1.0e-6, 1.0e-6, 1.0e-4, 1.0e-4, 1.0e-4
#R.diagonal() << 1.0e-12, 1.0e-12, 1.0e-12, 1.0e-3, 1.0e-3, 1.0e-3;

#data_file_03-15-19_21-31-25
#Q.diagonal() << 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-7, 1.0e-7, 1.0e-7, 1.0e-5, 1.0e-5, 1.0e-5;
#R.diagonal() << 1.0e-12, 1.0e-12, 1.0e-12, 1.0e-3, 1.0e-3, 1.0e-3;

#data_file_03-15-19_21-53-18
#   Q.diagonal() << 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-6, 1.0e-6, 1.0e-6;
#   R.diagonal() << 1.0e-12, 1.0e-12, 1.0e-12, 1.0e-3, 1.0e-3, 1.0e-3;

stehen geblieben!!!!!!!!!!!
#data_file_03-15-19_22-03-27
#   Q.diagonal() << 1.0e-8, 1.0e-8, 1.0e-8, 1.0e-7, 1.0e-7, 1.0e-7, 1.0e-6, 1.0e-6, 1.0e-6;
#   R.diagonal() << 1.0e-12, 1.0e-12, 1.0e-12, 1.0e-2, 1.0e-2, 1.0e-2;

#TUM colors rgb
blue = (0,0.3961,0.7412)
red = (0.7686,0.0275,0.1059)
green =(0,0.4863,0.1882)
orange =  (0.8902, 0.4471, 0.1333)
purple = (0.4118, 0.0314, 0.3529)
grey = (0.6118, 0.6157, 0.6235)
yellow = (0.9765, 0.7294, 0)
#turquoise = (0, 0.4667, 0.5412) 




def Plot_three_dim(time, data1, data2, label_1, label_2, title):
    f, axarr = plt.subplots(3,1, figsize=(6,8))
    f.suptitle(title, fontsize=20)
    coordinates = ["$_x$","$_y$","$_z$"]
    for idx,coordinate in enumerate(coordinates):
        axarr[idx].plot(time, data1[:,idx], c=blue, label= label_1)
        axarr[idx].plot(time, data2[:,idx], c=red, label= label_2)
        axarr[idx].set_xlim([0, time[-1]])
        axarr[idx].legend()
    axarr[2].set_xlabel("Elapsed time in $s$")
    plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    
def Plot_three_dim_one(time, data1, title):
    f, axarr = plt.subplots(3,1, figsize=(6,8))
    f.suptitle(title, fontsize=20)
    coordinates = ["$_x$","$_y$","$_z$"]
    for idx,coordinate in enumerate(coordinates):
        axarr[idx].plot(time, data1[:,idx], c=blue)
        axarr[idx].set_xlim([0, time[-1]])
#         axarr[idx].set_xlim([0,10])
        axarr[idx].set_ylim([-0.5,0.5])
    axarr[2].set_xlabel("Elapsed time in $s$")
    plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)


def PlotCompare(counter, time, x_1, x_2, y_1, y_2, z_1, z_2, title_1, title_2, title_3, label_1, label_2,title, savename):
	plt.figure(counter, figsize=(8,9))
	plt.title(title)
	plt.subplot(311)
	plt.tight_layout()
	plt.plot(time, x_1, c=blue, label= label_1)
	plt.plot(time, x_2, c=red, label= label_2)
	plt.title(title_1)
	plt.legend()
	plt.xlabel("Elapsed time in s")
	plt.subplot(312)
	plt.tight_layout()
	plt.plot(time, y_1, c=blue, label= label_1)
	plt.plot(time, y_2, c=red, label= label_2)
	plt.title(title_2)
	plt.legend()
	plt.xlabel("Elapsed time in s")
	plt.subplot(313)
	plt.tight_layout()
	plt.plot(time, z_1, c=blue, label= label_1)
	plt.plot(time, z_2, c=red, label= label_2)
	plt.xlabel("Elapsed time in s")
	plt.title(title_3)
	plt.legend()
	plt.savefig(savename)

```
